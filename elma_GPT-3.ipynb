```python
import json
import os
import openai

# question = "一番好きな食べ物は？"
question = "一番好きな人は？"

openai.api_key = os.getenv("OPENAI_API_KEY")

response = openai.Answer.create(
  search_model="ada", 
  model="curie",
  # search_model="davinci", 
  # model="davinci",
  question=question,
  file="file-4DuTleNtarWu5XsHUsz889B3",  # 直前のセルの出力にあるidの値に書き換える
  examples_context="私は地獄巡システムエンジニアリングで働いているのだ。", 
  examples=[["どこで働いているの？", "地獄巡システムエンジニアリングなのだ。"]],
  max_tokens=100,
  stop=["\n", "<|endoftext|>"],
)

print(json.dumps(response, ensure_ascii=False, indent=2))
```
